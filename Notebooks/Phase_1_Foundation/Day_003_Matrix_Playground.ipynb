# Torch and setup

```python
import torch
import numpy as np

print(f"Pytorch version: {torch.__version__}")

# Set a random seed for reproducability
torch.manual_seed(42)
```

# Defining Matrices

```python
# Create tensor with shape (2, 3)
A = torch.tensor([
    [1, 2, 3],
    [4, 5, 6],
], dtype=torch.float32)

# Create tensor with shape (3, 2)
B = torch.tensor([
    [7, 8],
    [9, 1],
    [2, 3]
], dtype=torch.float32)

print(f"Matrix A shape: {A.shape}")
print(f"Matrix B shape: {B.shape}")
print("\nMatrix A:\n", A)
print("\nMatrix B:\n", B)
```

# Matrix Multiplication (A x B)

```python
# Method 1: torch.matmul()
C = torch.matmul(A, B)

# Method 2: the @ operator (Standard python for Matrix Mult)
C_prime = A @ B

print(f"Result shape: {C.shape}") # should be (2, 2)
print("\nResult Matrix C:\n", C)

# Verify both methods are identical
assert torch.allclose(C, C_prime)
print("\nAssertion Passed: torch.matmul matches @ operator")
```

# Non Commutativity (B x A)

```python
D = B @ A

print(f"Result D (B @ A) Shape: {D.shape}")
print("\n Result Matrix D\n", D)
```

# Element-wise vs Matrix Multiplication



```python
#Let's create two square matrices to test this
X = torch.tensor([[1, 2], [3, 4]])
Y = torch.tensor([[5, 6], [7, 8]])

# 1. Element-wise multiplication (hadamard Product)
element_wise = X * Y
print("X * Y (Element-wise):\n", element_wise)

# 2. Matrix Multication
matrix_mult = X @ Y
print("\nX @ Y (Matrix Mult):\n", matrix_mult)
```
